```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df=pd.read_csv('supply_chain_data.csv')
```

```python
df
```

```python
df.describe()
```

```python
fig=px.scatter(df,x='Price',y='Revenue generated',
               color='Product type',
               hover_data=['Number of products sold'],
               trendline='ols')
fig.show()
```

```python
sale_data=df.groupby('Product type')['Number of products sold'].sum().reset_index()
sale_data
```

```python
pie_chart=px.pie(sale_data,values='Number of products sold',
                names='Product type',
                title='Sale by Product type',
                 hover_data=['Number of products sold'],
                hole=0.5

                )
pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

```python
df
```

```python
total_Revenue=df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
total_Revenue
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=total_Revenue['Shipping carriers']
           ,y=total_Revenue['Revenue generated']))
fig.update_layout(title='Total Revenue by Shipping carriers',
                 xaxis_title='Shipping carriers',yaxis_title='Revenue generated')
fig.show()
```

```python
revenue_chart=px.line(df,x='SKU',y="Revenue generated",title="Revenue generated by SKU")
revenue_chart.show()
```

```python
order_chart=px.line(df,x='SKU',y="Order quantities",title="Order quantitiesby SKU")
order_chart.show()
```

```python
shipping_cost=df.groupby('Shipping carriers')['Costs'].sum().reset_index()
shipping_cost
```

```python
fig=px.bar(shipping_cost,x="Shipping carriers",y='Costs',
           title='Shipping cost by Carriers')
fig.show()
```

```python
df
```

```python
Transportation_modes_cost=df.groupby('Transportation modes')['Costs'].sum().reset_index()
Transportation_modes_cost
```

```python
fig=px.pie(Transportation_modes_cost,values='Costs',names='Transportation modes',hole=0.5,title="Cost Distribution by Transportation Mode")
fig.show()
```

```python
defect_rate=df.groupby('Product type')['Defect rates'].mean().reset_index()
defect_rate
```

```python
fig=px.bar(defect_rate,x='Product type',y='Defect rates',title="defeect Rate Of Product")
fig.show()
```
